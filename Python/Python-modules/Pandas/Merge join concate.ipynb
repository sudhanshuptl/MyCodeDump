```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame(np.random.randn(4,4),['A','B','C','D'],['P','Q','R','S'])
df
```

```python
df1 = pd.DataFrame(np.random.randn(4,4),['A1','B1','C1','D1'],['P','Q','R','S'])
df1
```

```python
df2 = pd.DataFrame(np.random.randn(4,4),['A2','B2','C2','D2'],['P','Q','R','S'])
df2
```

# Concatenation

```python
pd.concat([df,df1,df2]) # since columns are common so concate is basicaly addtion of new row
```

```python
pd.concat([df,df1,df2],axis=1)  #forece fully adding as column
```

## Pandas have sql like joins also that merge like inner ,outer ,right or left join based on some common columns

```python
pd.merge(df1,df2,how='inner',on=['col1','col2'])
```
